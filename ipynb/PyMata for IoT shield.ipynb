## Introduction ##

In order to simplify all, special library for IoT shield was written

Use special library in order to simplify all. Download the library:
https://github.com/adamkaliszan/firmataPyClient
Launch en exemplary script: demo.py. Modify this script. Be familiar with functions for controlling 
the: 

- LEDs (leds objects are stored on leds array):
  - toggle() - change the state of the diode. If the diode was in on state the diode is disabled and vice versa
  - on() - switch on the diode
  - off() - switch off the diode
- Keys (keys are stored on key array)
  - read() - returns bool value
- Snsors(temperature and light)
  - read() - return the analogue value between 0 and 1023

The britness of the diodes 1-3 can be controlled using PWM. This function is not available for diode 4 and 5

## My Trick ##
I discovered small trick, that let me to launch PyMata-aio in Jypyter. You need to create cell with bash (not Python).

In this cell redirect the python script to python interpreter. In order to send a scrit to the stream use echo command.
There is a limitation, the symbols " " within a script neet to be precedet by escape sequence \".

Next use a pipe to redirect it to python 3.

Below there is an example.


```bash
#Downloading the library
if -ne IoTShield.py
then
    echo "Downloading the python module IoTShield.py"
    wget https://raw.githubusercontent.com/adamkaliszan/firmataPyClient/master/IoTShield.py
else
    echo "The IoTShield.py module is already downloaded, skipping downloaing"
fi

# Sending the python script to the output strem using echo command
echo "import IoTShield as AS

myBoard = AS.IoTShield(pwmMax=100)
print('Connection estabilished !!!')

ledNo=1
while True:
    myBoard.leds[ledNo].toggle()    
    myBoard.sleep(1)

    myBoard.leds[ledNo].toggle()
    myBoard.sleep(1)
        
    ledNo = 1 if ledNo == 5 else ledNo+1
    
    value = myBoard.SensorTemperature.read()    
    print('Temperature: {0}'.format(value))
    
    value = myBoard.SensorLight.read()    
    print('Light: {0}'.format(value))

" | python3
# The output stream was redirected to pipe and received by python3
```
